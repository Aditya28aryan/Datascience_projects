```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_rows',None)
```

```python
iris = pd.read_csv('Iris.csv')
```

```python
iris.head()
```

```python
sns.scatterplot(x='PetalLengthCm', y='PetalWidthCm',data=iris , hue='Species')
```

```python
iris.columns
```

```python
# here var(variable) is a new dataset where ID and species are dropped
var = iris.drop(['Id','Species'],axis=1)
```

```python
var
```

```python
wcss = []

for i in range(1,10) :
    
    from sklearn.cluster import KMeans
    kmeans = KMeans(n_clusters=i , random_state=0)
    kmeans.fit(var)
    wcss.append(kmeans.inertia_) # inertia is the wcss value 
```

```python
wcss
```

```python
plt.plot(range(1,10),wcss)
```

```python
kmeans = KMeans(n_clusters=3, random_state=0)
```

```python
clusters=kmeans.fit_predict(var)
```

```python
iris
```

```python
iris['clusters']=clusters
```

```python
iris
```

```python
#plot the clusters
fig , axs = plt.subplots(1,2, figsize=(12,5))
sns.scatterplot(x= 'PetalLengthCm', y= 'PetalWidthCm', data=iris , hue = 'clusters', ax= axs[0])
sns.scatterplot(x= 'PetalLengthCm', y= 'PetalWidthCm', data=iris , hue = 'Species', ax= axs[1])
```

```python
fig , axs = plt.subplots(1,2, figsize=(12,5))
sns.scatterplot(x= 'SepalLengthCm', y= 'SepalWidthCm', data=iris , hue = 'Species', ax= axs[0])
sns.scatterplot(x= 'SepalLengthCm', y= 'SepalWidthCm', data=iris , hue = 'clusters', ax= axs[1])
```
