# ROC and AUC

- ROC stands for **Receiver Output Characterics**
- AUC stands for **Area Under the Curve**

These curves help us assess the performance of classification models

- False Positive Rate = Of all the negative cases in the dataset how many the model incorrectly classified as positive

**FP / (TN + FP)**

- True Positive Rate = Of all the positive cases in the dataset how many the model correctly classified as positive

**TP / (TP+FN)**

- True positive Rate(TPR) is also called **sensitivity or recall**

- Sepcificity = It is the ability of the model to identify the negative cases correctly 

**TN / (TN + FP )**

1 - Specificity = False Positive Rate(FPR)

- The ROC curve is a graph between TPR and FPR . The area under the curve of this graph gives you a idea of the accuracy of the model 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
# x, y are two variable 
x , y = cancer.data , cancer.target
```

```python
# Divide the data into train and test split
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.2, random_state = 0)
```

```python
# Define Model
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
```

```python
log_reg.fit(x_train, y_train)
```

```python
y_pred = log_reg.predict_proba(x_test)[:,1]
```

```python
y_pred
```

```python
# Define ROC and AUC
from sklearn.metrics import roc_curve, auc

fpr, tpr, thresholds = roc_curve(y_test, y_pred)
```

```python
# Plot the Curve
plt.plot(fpr, tpr , color='blue')
plt.xlabel('FPR(False Positive Rate)', color ='r')
plt.ylabel('TPR(True Positive Rate)', color ='r')
plt.title('ROC Curve', color='r')
plt.show()
```

```python
# calculating the Area Under the curve
roc_auc = auc(fpr, tpr)
```

```python
print(roc_auc) # accuracy
```
