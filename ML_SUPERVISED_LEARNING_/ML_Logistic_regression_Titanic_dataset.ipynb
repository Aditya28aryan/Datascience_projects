```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

```python
titanic = pd.read_csv('titanic.csv')
```

```python
titanic
```

```python
titanic.info()
```

```python
titanic.isnull().sum()
```

```python
titanic['Age'] = titanic['Age'].fillna(titanic['Age'].mean())
```

```python
titanic['Age']
```

```python
titanic['Embarked']=titanic['Embarked'].fillna('S')
```

```python
titanic['Embarked']
```

```python
titanic.columns
```

```python
# All imp variables
```

```python
titanic_df = titanic[['Survived', 'Pclass', 'Sex', 'Age', 'SibSp',
       'Parch', 'Fare', 'Embarked']]
```

```python
titanic_df.info()
```

```python
titanic_df = pd.get_dummies(titanic_df,drop_first=True)
```

```python
titanic_df
```

```python
x = titanic_df[['Pclass', 'Age', 'SibSp', 'Parch', 'Fare', 'Sex_male', 'Embarked_Q','Embarked_S']]

y = titanic_df[['Survived']]
```

```python
x
```

```python
# Spliting the dataset in tgrain and test

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size =0.2 ,random_state=0)
```

```python
# scaling the dataset

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train[['Age', 'Fare']] = sc.fit_transform(x_train[['Age', 'Fare']])
x_test[['Age', 'Fare']] = sc.transform(x_test[['Age', 'Fare']])
```

```python
# fitting logistic regression model

from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(x_train,y_train)
```

```python
y_pred = classifier.predict(x_test)
```

```python
y_pred
```

```python
len(y_pred)
```

```python
# Check out Accuracy

from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score
acc = accuracy_score(y_pred, y_test)
cf = confusion_matrix(y_pred,y_test)
precision = precision_score(y_pred,y_test)
recall = recall_score(y_pred,y_test)
```

```python
print("Acuuracy:", acc)
print("precision:",precision)
print("recall:", recall)
```
