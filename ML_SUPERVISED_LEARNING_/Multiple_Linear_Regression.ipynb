```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('50_Startups.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df = pd.get_dummies(df,drop_first=True)
```

```python
x = df.drop(['Profit','State_Florida','State_New York'],axis=1)
y = df[['Profit']]
```

```python
x
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2,random_state=0)
```

```python
# Build the Linear Regression Model

from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
x_train
```

```python
x_test
```

```python
lr.score(x_train,y_train)
```

```python
import statsmodels.api as sm 

X2 = sm.add_constant(x_train)
est = sm.OLS(y_train,X2)
est2= est.fit()
est2.summary()
```

```python
y_pred = lr.predict(x_test)
```

```python
y_pred
```

```python
lr.intercept_
```

```python
lr.coef_
```

```python
42467+ 0.77*66051.52 + 0.033*182645.56 + 0.36*118148.20 + 0.86 -872 + 786
```

- **Sample model for show how it work's**

```python
## Making a prediction for a new data point

lr.predict([[100000,100000,200000]])
```

```python

from sklearn.metrics import mean_absolute_percentage_error, mean_absolute_error, mean_squared_error

rmse= np.sqrt(mean_absolute_error(y_test,y_pred))

mae = mean_absolute_error(y_test,y_pred)

mape=mean_absolute_percentage_error(y_test,y_pred)
```

```python
print("RMSE",rmse)
print("MAE",mae)
print("MAPE", mape)
```

```python
RMSE 86.68502557905033
MAE 7514.293659640612
MAPE 0.06449754085355504
```

```python
y_pred_train = lr.predict(x_train)
```

```python
y_pred_train
```

```python
from sklearn.metrics import mean_absolute_percentage_error, mean_absolute_error, mean_squared_error

rmse= np.sqrt(mean_absolute_error(y_train,y_pred_train))

mae = mean_absolute_error(y_train,y_pred_train)

mape=mean_absolute_percentage_error(y_train,y_pred_train)
```

```python
print("RMSE",rmse)
print("MAE",mae)
print("MAPE", mape)
```
